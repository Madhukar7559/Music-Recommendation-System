```python
import pandas as pd;
```

```python
x = pd.read_csv("testdata.csv");
lister = [];
x
```

```python
for i in range(0, x.shape[0]):
    text = x["track_url"][i].split("https://open.spotify.com");
    text = "https://open.spotify.com/embed".join(text) + "?utm_source=generator";
    lister.append(text);
len(lister)
```

```python
x["embed_code"] = lister;
```

```python
x.to_csv("actdata.csv", index=False);
```
